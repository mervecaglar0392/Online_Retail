"Does caffeine intake affect developer productivity?"

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=df["coffee_intake_mg"], y=df["sleep_hours"])
plt.xlabel("Caffeine Intake (mg)")
plt.ylabel("Sleep Hours")
plt.title("Caffeine Intake vs Sleeping Hours")
plt.show()
```

```python
sns.scatterplot(x=df["coffee_intake_mg"], y=df["hours_coding"])
plt.xlabel("Caffeine Intake (mg)")
plt.ylabel("Hours_Coding")
plt.title("Caffeine Intake vs Coding Hours")
plt.show()
```

```python
sns.scatterplot(x=df["coffee_intake_mg"], y=df["distractions"])
plt.xlabel("Caffeine Intake (mg)")
plt.ylabel("distractions")
plt.title("Caffeine Intake vs Distractions")
plt.show()
```

Relationship Between Sleep Hours and Task Success

```python
sns.scatterplot(x=df["sleep_hours"], y=df["task_success"])
plt.xlabel("sleep_hours")
plt.ylabel("task_success")
plt.title("Relationship Between Sleep Hours and Task Success")
plt.show()
```

"Relationship Between Coding Hours and Ai Usage Hours"

```python
sns.scatterplot(x=df["hours_coding"], y=df["ai_usage_hours"])
plt.xlabel("hours_coding")
plt.ylabel("ai_usage_hours")
plt.title("Relationship Between Coding Hours and Ai Usage Hours")
plt.show()
```

```python
sns.scatterplot(x=df["hours_coding"], y=df["cognitive_load"])
plt.xlabel("hours_coding")
plt.ylabel("cognitive_load")
plt.title("Relationship Between Coding Hours and Cognitive Load")
plt.show()
```

```python
sns.scatterplot(x=df["hours_coding"], y=df["bugs_reported"])
plt.xlabel("hours_coding")
plt.ylabel("bugs_reported")
plt.title("Relationship Between Coding Hours and Bugs Reported")
plt.show()
```

```python
sns.scatterplot(x=df["hours_coding"], y=df["task_success"])
plt.xlabel("hours_coding")
plt.ylabel("task_success")
plt.title("Relationship Between Coding Hours and Task Success")
plt.show()
```

```python
sns.scatterplot(x=df["bugs_reported"], y=df["commits"])
plt.xlabel("bugs_reported")
plt.ylabel("commits")
plt.title("Relationship Between commits and bugs reported")
plt.show()
```
