| Column Name          | Ideal `dtype`          | Explanation                                                              |
| -------------------- | ---------------------- | ------------------------------------------------------------------------ |
| **Transaction ID**   | `object` or `string`   | Usually a unique identifier, typically non-numeric                       |
| **Item**             | `category` or `object` | Categorical data; if there are few unique items, `category` is efficient |
| **Quantity**         | `int64` or `float64`   | Number of items sold; integer preferred unless missing values exist      |
| **Price Per Unit**   | `float64`              | Unit price; decimal numbers                                              |
| **Total Spent**      | `float64`              | Usually calculated as Quantity  Price Per Unit                          |
| **Payment Method**   | `category` or `object` | Payment type; categorical with limited unique values                     |
| **Location**         | `category` or `object` | Store or branch location; categorical if few unique values               |
| **Transaction Date** | `datetime64[ns]`       | Date and time information; use datetime type for time series analysis    |


```python
df.dtypes
```

```python
df['Quantity'] = df['Quantity'].astype(str).str.strip()
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')

```

```python
print(df['Price Per Unit'].dtype)
```

```python
df['Price Per Unit'] = df['Price Per Unit'].fillna('').astype(str).str.replace('[\$,]', '', regex=True).str.strip()
```

```python
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
```

```python
print(df['Price Per Unit'].dtype)
```

```python
print(df['Total Spent'].dtype)
print(df['Total Spent'].unique()[:10])  # Show first 10 unique values to check for issues
```

```python
df['Total Spent'] = df['Total Spent'].astype(str).str.replace(r'[\$,]', '', regex=True).str.strip()
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
```

```python
print(df['Price Per Unit'].dtype)
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```
